```python
import pandas as pd
```

```python
data = pd.read_csv('CSE_student_performances.csv')
```

```python
# 1
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# 2
data.drop('LikePresentation', axis=1, inplace=True)
```

```python
data.head(0) # проверяем
```

```python
# 3
data.rename(columns={
    'AcademicPerformance': 'Academic Performance',
    'TakingNoteInClass': 'Taking Note In Class',
    'DepressionStatus': 'Depression Status',
    'FaceChallangesToCompleteAcademicTask': 'Face Challanges To Complete Academic Task',
    'LikePresentation': 'Like Presentation',
    'SleepPerDayHours': 'Sleep Per Day Hours',
    'NumberOfFriend': 'Number Of Friend',
    'LikeNewThings': 'Like New Things'
}, inplace=True)
```

```python
# 4
data = data.drop_duplicates().dropna()
```

```python
# 5 Выявить среднее количество сна студентов
average_sleep = data['Sleep Per Day Hours'].mean()
print(f'{average_sleep:.2f} ')

# Тут 6.73 от общего числа студентов
```

```python
average_sleep_excellent = data[data['Academic Performance'] == 'Excellent']['Sleep Per Day Hours'].mean()
print(f' {average_sleep_excellent:.2f}')

# Вот тут показано по excellent среднее значение сна
```

```python
# 6 Вывести таблицу, где только студенты с оценками "excellent" и посчитать среднее значение сна

excellent_students = data[data['Academic Performance'] == 'Excellent']
avg_sleep_excellent = excellent_students['Sleep Per Day Hours'].mean()
excellent_students['Avg Sleep (Excellent)'] = avg_sleep_excellent

excellent_students

# 7.66 только по excellent студентам, 7.67 показывает так как мы использовали округление после нуля (.2f)
```

```python
# 7 Вывести таблицу, где только студенты с депрессией - yes и посчитать среднее значение их сна

depressed_students = data[data['Depression Status'] == 'Yes']
avg_sleep_depression = depressed_students['Sleep Per Day Hours'].mean()
depressed_students['Avg Sleep (Depression)'] = avg_sleep_depression

depressed_students
```
